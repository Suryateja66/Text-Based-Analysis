# Pre-processing pipeline

```python
# Importing the regex module for regular expression operations.
import regex as re
```

```python
# Defining a Function to Pull Words
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

# Tokenizing the Given Text

```python
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

# Using nltk Package

```python
# Importing the Natural Language Toolkit (nltk) library.
import nltk
# Downloading stopw
nltk.download('stopwords')
# Creating a set of English stopwords from the nltk stopwords corpus.
stopwords=set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
# Function to remove stop words
def remove_stop(tokens):
    return[t for t in tokens if t.lower() not in stopwords]
```

```python
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

# Making a PipeLine

```python
# Creating a Pipeline
pipeline = [str.lower, tokenize, remove_stop]
```

```python
# Defining a Function to Run the Pipeline
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# Runing PipeLine

```python
# running pipeline
df['tokens'] = df['Division'].apply(prepare, pipeline=pipeline)
```

# Counting Tokens

```python
# Count the tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Word Frequency Analysis

```python
# Importing the Counter class from the collections module.
from collections import Counter
```

```python
tokens = tokenize("She likes my cats and my cats like my sofa.")
```

```python
# Counting number of times words are repeating Using "Counter"
testcount= Counter(tokens)
```

```python
print(testcount)
```

# Adding New Data

```python
# Adding new data to the current count
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

# Counting Word Repetations

```python
# Printing the 5 Words Repeated the Most
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
# Putting the word frequency in its own dataframe for analysis 
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq'])
```

```python
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

# Sorting Based on Frequency

```python
# Sort and display
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

# Printing Statistical Data

```python
#Statistical data
freq_df.describe()
```

# Bar chart

```python
# Plotting a bar graph for most frequent words
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
